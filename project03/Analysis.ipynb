```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\ellie\\OneDrive\\Desktop\\project03\\results.csv")
df.head(10)
```

# Data Visualization 1

```python
sns.set_style("white")
plt.figure(figsize = (8, 6))
df1 = df.groupby('ticker').max()['highest_hourly_stock_price'].to_frame()
df1.reset_index(inplace=True)

ax = sns.catplot(x='ticker',y='highest_hourly_stock_price',data=df1, kind='bar', color = "midnightblue")

plt.title("Highest Price", fontsize = 14, weight = "bold")
plt.xlabel("Ticker")
plt.ylabel("Price")
sns.despine()
plt.show()
```

# Data Visualization 2

```python
df = df[['ticker','highest_hourly_stock_price','hour']]

for i in df['ticker'].unique():
    plot = df[df['ticker'] == i]
    plot.set_index('hour').plot(kind='line')
    plt.title(f'Hourly Highest Stock Price Change of {i}', fontsize = 13, fontweight = "bold")
    plt.xlabel("Hour", fontsize = 12, fontweight = "bold")
    plt.ylabel("Hourly Highest Price", fontsize = 12, fontweight = "bold")
    plt.show()
```
